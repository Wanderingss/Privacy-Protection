```python
import numpy as np
import matplotlib.pyplot as plt

# 设置随机种子以便复现
np.random.seed(42)

# 生成时间序列
time_steps = 100
time = np.arange(time_steps)

# 生成真实位置数据（例如，一个物体以0.5的速度移动，并添加一些随机波动）
true_position = 0.5 * time + np.random.normal(0, 2, time_steps)

# 差分隐私参数
epsilon = 1
sensitivity = 1.0  # 假设敏感度为1

# 计算拉普拉斯噪声的尺度参数
b = sensitivity / epsilon

# 生成拉普拉斯噪声
laplace_noise = np.random.laplace(0, b, time_steps)

# 添加噪声到真实位置
noisy_position = true_position + laplace_noise

# 移动平均窗口大小
window_size = 5

# 计算移动平均
def moving_average(data, window_size):
    ret = np.cumsum(data, dtype=float)
    ret[window_size:] = ret[window_size:] - ret[:-window_size]
    return ret[window_size - 1:] / window_size

# 对噪声位置进行移动平均
estimated_position = moving_average(noisy_position, window_size)

# 由于移动平均会缩短数据长度，我们需要调整时间轴
estimated_time = time[window_size - 1:]

# 可视化真实位置、噪声位置和估计位置
plt.figure(figsize=(12, 6))
plt.plot(time, true_position, label='True Position', color='blue')
plt.plot(time, noisy_position, label='Noisy Position (ε=1.0)', color='orange', alpha=0.7)
plt.plot(estimated_time, estimated_position, label='Estimated Position (Moving Average)', color='green')
plt.title('True Position, Noisy Position, and Estimated Position')
plt.xlabel('Time Step')
plt.ylabel('Position')
plt.legend()
plt.grid(True)
plt.show()
```
